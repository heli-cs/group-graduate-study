### OpenCVのHaar Cascadesを使った顔検出

Haar特徴ベースのCascade型分類器を使った物体検出はPaul ViolaとMichael Jonesが2001年に”Rapid Object Detection using a Boosted Cascade of Simple Features”という論文で発表した効率過的物体検出手法です。

OpenCVは学習機と検出器の両方を提供しています。ここには検出の部分を行います。

先ずは、`opencv/data/haarcascades`フォルダ内に保存されたXMLファイルから検出器を読み込みます。

```
face_cascade = cv2.CascadeClassifier()
if not face_cascade.load('/usr/local/lib/python3.6/dist-packages/cv2/data/haarcascade_frontalface_alt.xml'):
  print('--(!)Error loading face cascade')
  exit(0)
```

次は、入力画像(もしくは動画像)をグレースケールモードで読み込みます。



```
from google.colab.patches import cv2_imshow


filepath = "/content/drive/My Drive/OpenCV_test.png"
img = cv2.imread(filepath)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

最後は、入力画像に対して顔の位置が出力されます。

```
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x, y, w, h) in faces:
  print("Left-top point x:", x)
  print("Left-top point y:", y)
  print("Width:", w)
  print("Heeight:", h)

# cv2_imshow(img)
```
